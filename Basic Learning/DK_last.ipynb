## Difference between radioactive decay and estimating Pi logic

**Instead of asking**

*Is this point inside the sphere?*

we ask

### *Does this atom decay?*
> `decays = np.sum(np.random.random(N) < lam)`

Each atom is like one random point.

Suppose there are 10 atoms.
We generate

`np.random.random(10)`

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)  # For reproducibility

N0 = 1000  # Number of initial atoms
lmd = 0.1  
steps = 100  # Number of time steps

DK_MC = np.zeros(steps)

N=N0

for i in range(steps):
    atoms = np.random.random(N)
    decay = np.sum(atoms < lmd)
    N -= decay
    DK_MC[i] = N

t = np.arange(steps)
DK_analytic = N0 * np.exp(-lmd * t)

# 1. Plot the Monte Carlo simulation results
plt.figure(figsize=(10, 6))
plt.plot(t, DK_MC, label='Monte Carlo')
plt.plot(t, DK_analytic, label='Analytic')
plt.xlabel('Time')
plt.ylabel('Number of Atoms')
plt.title('Radioactive Decay Simulation')
plt.legend()
plt.show()  
```
